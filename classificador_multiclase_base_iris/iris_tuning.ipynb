# importando as libs 

```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

# Carregando o dataset íris e separando os previsores e as classes

```python
base = pd.read_csv("./iris.csv")
```

```python
previsores = base.iloc[:, :4].values
classes = base.iloc[:, 4].values

previsores.shape
```

# Criando a estrutura da Rede Neural

```python
def criarRede(optimizer, kernel_initializer, activation, neurons):
    classificador = Sequential()
    classificador.add(Dense(units=neurons, activation=activation,
                            kernel_initializer=kernel_initializer, input_dim=4))
    classificador.add(Dropout(0.15))
    classificador.add(Dense(units=neurons, activation=activation,
                           kernel_initializer=kernel_initializer))
    classificador.add(Dropout(0.15))
    classificador.add(Dense(units=3, activation="softmax"))
    
    classificador.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy",
                         metrics=["accuracy"])
    return classificador
    
```

# Instanciando a Rede Neural

```python
classificador = KerasClassifier(build_fn=criarRede)
```

# Criando os parâmetros para testar a Rede Neural

```python
parametros = {'batch_size': [8, 15],
              'epochs': [80, 100],
              'optimizer': ['adam', 'RMSprop'],
              'kernel_initializer': ['random_uniform', 'GlorotUniform'],
              'activation': ['relu', 'elu'],
              'neurons': [8, 4]}
```

# Criando o Grid Search e rodando os avaliador

```python
grid_search = GridSearchCV(estimator=classificador,
                           param_grid=parametros,
                           cv=5)
```

# Iniciando os testes

```python
grid_search.fit(previsores, classes)
```

# Obtendo os melhores parâmetros

```python
melhores_parametros = grid_search.best_params_
melhor_precisao = grid_search.best_score_
```
