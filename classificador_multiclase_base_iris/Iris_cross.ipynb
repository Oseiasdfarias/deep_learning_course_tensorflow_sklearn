# Cross Validation

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
base = pd.read_csv("./iris.csv")
```

# Pré-Processamento dos dados

### Separando os atributos previsores e a classe

```python
previsores = base.iloc[:, :4].values
classe = base.iloc[:, 4].values

print(previsores.shape)
print(classe.shape)
```

### modificando os atributos da classe em numeros

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
classe = labelencoder.fit_transform(classe)
```

### Agora vamos transformar as classes em valores que podem ser entendido pela função softmax
exemplo: 
    [2
     3
     5]
     
para que a função softmax entenda essas classes, temos que reescrevelas da seguinte forma


[[100]
 [010]
 [001]]
 
 

```python
classe_dummy = np_utils.to_categorical(classe)
```

# Criar uma função para criar uma rede neural com keras e tensorflow

```python
def criarRede():
    classificador = Sequential()
    classificador.add(Dense(units=4, activation="relu", input_dim=4))
    classificador.add(Dense(units=3, activation="softmax"))
    
    classificador.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["categorical_accuracy"])
    
    return classificador
```

# Criando o rede neural

```python
classificador = KerasClassifier(build_fn=criarRede,
                                epochs=800,
                                batch_size=10)
```

```python
resultados = cross_val_score(estimator=classificador,
                            X=previsores,
                            y=classe,
                            cv=10,
                            scoring="accuracy")
```

```python
media = resultados.mean()
desvio = resultados.std()
print(f"Média de Accuracy: {media}")
print(f"DesvIO Padrão: {desvio}")
```

# Salvamdo Estrutura da Rede Neural Artificial no formato json

```python
classificador_json = classificador.to_json()

# Criando um arquivo json e escrevendo nele o modelo treinado
with open("classificador_iris.json", "w") as json_file:
    json_filefile.write(classificador_json)
    
```
