# base de dados iris 


```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```python
base = pd.read_csv("./iris.csv")
```

```python
base.head()
```

```python
base.shape
```

### iloc seleciona as linhas e colunas, values transforma no formato numpy

```python
previsores = base.iloc[:, 0:4].values
```

```python
classe = base.iloc[:, 4].values
```

# Pré- Processando os dados

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lebal_encoder = LabelEncoder()

classe = lebal_encoder.fit_transform(classe)  # transformando de valores de classes em valores numéricos
classe_dummy = np_utils.to_categorical(classe)  # transformando em valores binários

# print(classe_dummy)
```

# Criando o comjunto de treinamento e o conjunto de teste

```python
from sklearn.model_selection import train_test_split
```

```python
prev_train, prev_test, class_train, class_test = train_test_split(previsores, classe_dummy, test_size=0.25)
```

# Construção da estrutura da rede neural

```python
classificador = Sequential()
```

```python
classificador.add(Dense(units=4, activation="relu", input_dim=4))
classificador.add(Dense(units=4, activation="relu"))
classificador.add(Dense(units=3, activation="softmax"))

classificador.compile(optimizer="adam",
                      loss="categorical_crossentropy",
                      metrics=["categorical_accuracy"])
```

```python
classificador.fit(prev_train, class_train, batch_size=10, epochs=1000)
```

```python
resultado = classificador.evaluate(prev_test, class_test)
```

```python
resultado
```

```python
previsoes = classificador.predict(prev_test)
```

```python
previsoes = (previsoes > 0.5)
```

```python
import  numpy as np
classe2 = [np.argmax(t) for t in class_test]

# print(classe2)

previsoes2 = [np.argmax(t) for t in previsoes]

# print(previsoes2)
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(previsoes2, classe2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn

df_cm = pd.DataFrame(matrix, index = [i for i in "ABC"],
              columns = [i for i in "ABC"])
plt.figure(figsize = (10,7))
sn.heatmap(df_cm, annot=True, cmap="Greens")
```
